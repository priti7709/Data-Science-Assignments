```python
# Train Test Split:-
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = glass_data.drop('Type', axis = 1)
y = glass_data['Type']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.30,random_state = 42)
```

```python
X_train.shape,X_test.shape
```

```python
y_test.shape,y_train.shape
```

```python
# KNN (K Neighrest Neighbour Classifier):-
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
k_range = range(1, 40)
k_scores = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors = k)
    train_scores = cross_val_score(knn, X_train, y_train, cv = 5)
    k_scores.append(train_scores.mean())

plt.figure(figsize = (10,6))
plt.plot(k_range, k_scores,marker = "o",color = 'blue')
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.show()
```

```python
#Model :-
```

```python
model = KNeighborsClassifier(n_neighbors = 2)
model
```

```python
model_prediction = model.fit(X_train,y_train).predict(X_train)
model_prediction
```

```python
model_accuracy = model.score(X_test, y_test)
model_accuracy
print('Model accuracy is:',model_accuracy)
```

```python
# Plot Confusion Matrix:-
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm_pred = model.predict(X_test)
cm = confusion_matrix(y_test, cm_pred)
cm
```

```python
prediction_df = pd.DataFrame({'Actual' : y_test, 'Predicted' : cm_pred})
prediction_df.head()
```

```python
plt.figure(figsize = (11,7))

sns.heatmap(cm, annot = True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
```

```python
#Grid Search for Algorithm Tuning:-
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
n_neighbors = np.array(range(1,40))
param_grid = dict(n_neighbors = n_neighbors)
```

```python
model = KNeighborsClassifier()
model
```

```python
grid = GridSearchCV(estimator = model, param_grid = param_grid)
grid.fit(X_train, y_train)
```

```python
print('The Best Score Accuracy :',grid.best_score_)
print('The Best Parameter  :',grid.best_params_)
```

```python
#Result :
  # a) For KNN Model Accuracy of Animals  : 0.6615

  # b) The Grid Best Score Accuracy       : 0.6574
```

```python
#Train Test Split :-
```

```python
from sklearn.model_selection import train_test_split
```

```python
animals_data_df = animals_data.drop('animal name', axis = 1)
animals_data_df.head()
```

```python
X = animals_data_df.drop('type', axis = 1)
y = animals_data_df['type']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state = 0)
```

```python
X_train.shape,X_test.shape
```

```python
y_test.shape,y_train.shape
```

#KNN (K Neighrest Neighbour Classifier)

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
# choose k between 1 to 41
k_range = range(1, 31)
k_scores = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors = k)
    train_scores = cross_val_score(knn, X_train, y_train, cv = 5)
    k_scores.append(train_scores.mean())

plt.figure(figsize = (12,8))
plt.plot(k_range, k_scores,marker = "o")
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.show()
```

```python
#Model :-
```

```python
model = KNeighborsClassifier(n_neighbors = 5)
model
```

```python
model_pred = model.fit(X_train,y_train).predict(X_train)
model_pred
```

```python
model_accuracy = model.score(X_test, y_test)
model_accuracy
print('Model accuracy is:',model_accuracy)
```

```python
#Plot Confusion Matrix:-
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm_pred = model.predict(X_test)
cm = confusion_matrix(y_test, cm_pred)
cm
```

```python
pred_df = pd.DataFrame({'Actual' : y_test, 'Predicted' : cm_pred})
pred_df.head()
```

```python
plt.figure(figsize = (10,6))

sns.heatmap(cm, annot = True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
```

```python
#Grid Search for Algorithm Tuning:-
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
n_neighbors = np.array(range(1,40))
param_grid = dict(n_neighbors = n_neighbors)
```

```python
model = KNeighborsClassifier()
model
```

```python
grid = GridSearchCV(estimator = model, param_grid = param_grid)
grid.fit(X_train, y_train)
```

```python
print('The Best Score Accuracy :',grid.best_score_)
print('The Best Parameter  :',grid.best_params_)
```
